```python
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import SGD, Adam

import numpy as np

print(tf.__version__)
```

```python
xData = np.array([
  [1,2,0],[5,4,3],[1,2,-1],[3,1,0],[2,4,2],[4,1,2],[-1,3,2],[4,3,3],[0,2,6],[2,2,1],[1,-2,-2],[0,1,3],[1,1,3],[0,1,4],[2,3,3]
  ])
tData = np.array([-4,4,-6,3,-4,9,-7,5,6,0,4,3,5,5,1])

print('xData_shape = ',xData.shape,' , tData_shape = ',tData.shape)
```

```python
model = Sequential()

model.add(Dense(1,input_shape=(3,) ,activation="linear")) 
```

```python
model.compile(optimizer=SGD(learning_rate=1e-2), loss='mse')

model.summary()
```

```python
hist = model.fit(xData, tData, epochs=1000)
```

```python
result = model.predict(np.array([-3.1,3.0,3.5,15.0,20.1]))

print(result)
```
